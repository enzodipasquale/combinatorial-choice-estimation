```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load the MIF/MID file
map_data = gpd.read_file("/Users/Enzo/Desktop/JMP/bta_map_data/bta.mif")

# Create a larger figure manually
fig, ax = plt.subplots(figsize=(30, 20), dpi=200)

# Plot the geometries on the custom axes
map_data.plot(edgecolor="black", color='none', ax=ax)

# Remove axes and add a title
ax.axis("off")
plt.title("Colored Map of Areas", fontsize=30)
plt.show()


```
